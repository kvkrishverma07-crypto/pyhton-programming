1. Create:
- Class `Person` → attributes: `name`
- Class `Student` (inherits Person) → attribute: `course`
- Class `GraduateStudent` (inherits Student) → attribute: `degree`

👉 Task:

- Create object of `GraduateStudent`
- Access all attributes
- Use proper `super()` chaining

````python
class Person:
    def __init__(self, name):
        self.name=name

class Student(Person):
    def __init__(self, name, course):
        super().__init__(name)
        self.course=course

class GraduateStudent(Student):
    def __init__(self, name, course, degree):
        super().__init__(name, course)
        self.degree=degree
    
    def display(self):
        return f"Name: {self.name}, Course: {self.course}, Degree: {self.degree}"
    
s1=GraduateStudent("krish", "Python", "BCA")
print(s1.display())
````

2. Create:
- Class `A` → method `show()` prints `"Class A"`
- Class `B(A)` → overrides `show()`
- Class `C(B)` → overrides `show()`

👉 Task:

Call `show()` using object of `C`
Print MRO using:  
```python
print(C.mro())
```

````python
class A:
    def show(self):
        print("Class A")

class B(A):
    def show(self):
        print("Class B")

class C(B):
    def show(self):
        print("Class C")

obj=C()
obj.show()
print(C.mro())
````

3. Create:
- Class `Father` → method `skills()` → "Gardening"
- Class `Mother` → method `skills()` → "Cooking"
- Class `Child(Father, Mother)`

👉 Task:

- Call `skills()` using Child object
Observe which method runs

````python
class Father:
    def skills(self):
        print("Gardening")

class Mother:
    def skills(self):
        print("Cooking")

class Child(Father, Mother):
    pass

obj=Child()
obj.skills()
````

4. Modify previous example:

👉 In `Child`, override `skills()` and:

- call both parent methods using `super()` or direct calls

Expected:

- Output both "Gardening" and "Cooking"

````python
class Child(Father, Mother):
    def skills(self):
        Father.skills(self)
        Mother.skills(self)

obj=Child()
obj.skills()
````

5. Create:

- Class `Device` → method `power_on()`
- Class `Camera` → method `take_photo()`
- Class `Phone` → method `make_call()`

👉 Create class `Smartphone(Device, Camera, Phone)`

👉 Tasks:

- Use all methods from different parents
- Demonstrate how multiple inheritance combines features

````python
class Device:
    def power_on(self):
        print("Switching On!!")

class Camera:
    def take_photo(self):
        print("Clicked :)")

class Phone:
    def make_call(self):
        print("Ringing...")

class Smartphone(Device, Camera, Phone):
    pass

obj=Smartphone()
obj.power_on()
obj.take_photo()
obj.make_call()
````
