1. Create a parent class `Person`:
- attributes: `name`, `age`
- method: `display_info()`  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Create a child class `Student` that:
- inherits from `Person`
- adds attribute: `course`

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Create a student object and display all details

```python
class Person:
    def __init__(self, name, age):
        self.name=name
        self.age=age
    
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Student(Person):
    def __init__(self, name, age, course):
        self.name=name
        self.age=age
        self.course=course

s1=Student("krish", 18, "Python")
s1.display_info()
print("Course: ", s1.course)
```

2. Modify the `Student` class:
- Use `super()` to initialize parent attributes

ЁЯСЙ Without super(), your design is incomplete

```python
class Student(Person):
    def __init__(self, name, age, course):
        super().__init__(name, age)
        self.course=course

s2=Student("krish", 19, "DSA")
s2.display_info()
print("Course:", s2.course)
```

3. Create:
- Parent class `Animal` with method `sound()` тЖТ prints `"Some generic sound"`
- Child classes:
    - `Dog` тЖТ `"Bark"`
    - `Cat` тЖТ `"Meow"`

ЁЯСЙ Override the method in child classes

```python
class Animal:
    def sound(self):
        print("Some generic sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

d=Dog()
c=Cat()
a=Animal()

d.sound()
c.sound()
a.sound()
```

4. Using the `Animal`, `Dog`, `Cat` classes:

ЁЯСЙ Create a list of different animal objects  
ЁЯСЙ Loop through them and call `sound()`

```python
animals=[Dog(), Cat(), Animal()]
for animal in animals:
    animal.sound()
```

5. Create:
- Parent class `Vehicle`:
    - method `start_engine()`
- Child classes:
    - `Car`
    - `Bike`

ЁЯСЙ Each should override `start_engine()` with different messages

ЁЯСЙ Create objects and call method for each

```python
class Vehicle:
    def start_engine(self):
        print("Starting Engine...")

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started using ignition")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine started using self-start")

v1=Car()
v2=Bike()
v3=Vehicle()

v1.start_engine()
v2.start_engine()
v3.start_engine()
```
