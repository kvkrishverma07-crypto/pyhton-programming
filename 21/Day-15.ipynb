1. Create class `Counter`:

- attribute: `count` (start = 0)

👉 Methods:

- `increment()` → increases count
- `decrement()` → decreases count (not below 0)
- `show()`

👉 Constraint:

- count should never go below 0

```python
class Counter:
    def __init__(self):
        self.count=0
    
    def increment(self):
        self.count+=1
        return "Count Increased!!!"
    
    def decrement(self):
        if self.count==0:
            return "Can not go below 0!!"
        else:
            self.count-=1
            return"Count decreased!!"
    
    def show(self):
        return f"Current Count: {self.count}"

c1=Counter()
print(c1.increment())
print(c1.decrement())
print(c1.decrement())
print(c1.increment())
print(c1.show())
```

2. Create class `Quiz`:

- attributes:
    - `question`
    - `answer`
    - `score` (initially 0)

👉 Method:

- `check_answer(user_answer)`
    - if correct → increase score
    - else → print wrong

👉 Task:

- Ask same question multiple times

```python
class Quiz:
    def __init__(self, question, answer):
        self.question=question
        self.answer=answer
        self.score=0
    
    def check_answer(self, answer):
        if answer==self.answer:
            print("Correct Answer!!")
            self.score+=1
        else:
            print("Wrong Answer!!")

Q1=Quiz("Capital of India?", "Delhi")
Q1.check_answer("Mumbai")
Q1.check_answer("Delhi")
```

3. Create class `File`:

- attributes:
    - `name`
    - `is_open` (False initially)

👉 Methods:

- `open_file()`
- `close_file()`
- `status()`

👉 Task:

- Try opening twice
- Try closing twice

```python
class File:
    def __init__(self, name):
        self.name=name
        self.is_open=False
    
    def open_file(self):
        if self.is_open:
            print("File already opened")
        else:
            self.is_open=True
            print("File Opened!!")
    
    def close_file(self):
        if not self.is_open:
            print("File already closed!!")
        else:
            self.is_open=False
            print("File Closed!!")
    
    def status(self):
        print("Current Status: ", self.is_open)

f1=File("ABC")
f1.open_file()
f1.open_file()
f1.close_file()
f1.close_file()
f1.status()
```

4. Create class `Product`:

- attributes: `name`, `price`  

👉 Method:

- `apply_discount(percent)`
    - reduce price accordingly
    - percent must be valid (0–100)

👉 Task:

- Apply valid + invalid discount

```python
class Product:
    def __init__(self, name, price):
        self.name=name
        self.price=price
    
    def apply_discount(self, percent):
        if percent<0 or percent>100:
            print("Percent must be valid (0-100)")
        else:
            print("Discounted Price:", self.price-(self.price*percent)/100)

p1=Product("Mobile", 22000)
p1.apply_discount(50)
p1.apply_discount(110)
```
