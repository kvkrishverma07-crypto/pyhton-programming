1. Create a class `Employee`:  
- instance variables: `name`, `salary`
- class variable: `company = "ABC Corp"`  
  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Create 2 objects and:  
- Print their details
- Change company name using one object

```python
class Employee:
    company="ABC Corp"
    
    def __init__(self, name, salary):
        self.name=name
        self.salary=salary
    
    def details(self):
        print(f"Name: {self.name}, Salary: {self.salary}, Company: {self.company}")

e1=Employee("krish", 50000)
e2=Employee("krrish", 80000)

e1.details()
e2.details()

e2.company="XYZ Corp"
print("\n---Updated Details---\n")
e1.details()
e2.details()
```

2. Create a class `Student`:
- attribute: `name`
- class variable:  `total_students` (initially 0)  
  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Every time a new object is created:
- Increment `total_students`  
  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Print total students after creating 3 objects

```python
class Student:
    total_students=0
    
    def __init__(self, name):
        self.name=name
        Student.total_students+=1

s1=Student("krish")
print(f"Current strength: {Student.total_students}")

s2=Student("rahsvvg")
print(f"Current Strength: {Student.total_students}")
```

3. Create a class `Calculator`:  
- methods:
    - `add(a, b)`
    - `multiply(a, b)`
    - `calculate(a, b)` → should:
        - call both methods
        - print sum and product

```python
class Calculator:
    def __init__(self, a, b):
        self.a=a
        self.b=b
    
    def add(self):
        return self.a+self.b
    
    def multiply(self):
        return self.a*self.b
    
    def calculate(self):
        print(f"Addition: {self.add()}, Multiplication: {self.multiply()}")
        
calc1=Calculator(3, 4)
calc1.calculate()
```

4. Create a class `LibraryBook`:
- attributes: `title`, `author`, `is_available` (default = True)
- methods:
    - `borrow_book()` → only if available
    - `return_book()`
    - `display_status()`

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Simulate:  

- Borrowing the same book twice
- Returning it

```python
class LibraryBook:
    def __init__(self, title, author):
        self.title=title
        self.author=author
        self.is_available=True
        
    def borrow_book(self):
        if self.is_available:
            self.is_available=False
            print(f"Borrowed: {self.title}")
        else:
            print(f"Book {self.title} not available")
    
    def return_book(self):
        self.is_available=True
        print(f"Returned: {self.title}")
    
    def display_status(self):
        Status="Available" if self.is_available else "Not Available"
        print(f"Title: {self.title}, Author: {self.author}, Status: {Status}")

book1=LibraryBook("idk", "ABC")
book2=LibraryBook("uknowme", "Someoneyouknow")

book1.borrow_book()
book1.display_status()
book2.display_status()
```
