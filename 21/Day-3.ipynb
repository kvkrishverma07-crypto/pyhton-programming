1. Create a class `Car`:
- attributes: `brand`, `model`
- implement `__str__()` so that printing the object shows:  
`"Brand: Toyota, Model: Fortuner"`

```python
class Car:
    def __init__ (self, brand, model):
        self.brand=brand
        self.model=model
    
    def __str__(self):
        return f"Brand: {self.brand}, Model: {self.model}"

c1=Car("Toyota", "Fortuner")
print(c1)
```

2. Create a class `Student`:
- attributes: `name`, `marks`

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Tasks:

- Create 2 objects
- Compare them using ==

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Then implement `__eq__()` such that:

- Two students are equal if their marks are the same

```python
class Student:
    def __init__(self, name, marks):
        self.name=name
        self.marks=marks
    
    def __eq__(self, other):
        return self.marks==other.marks

s1=Student("krish", 82)
s2=Student("Harsh", 82)
s3=Student("ravi", 89)

print(s1==s2)
print(s2==s3)
```

3. Create a class `Book`:
- attributes: `title`, `author`

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Implement:

- `__repr__()` that returns a string like:  
`"Book('Atomic Habits', 'James Clear')"`

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Then:

- Create a list of book objects
- Print the list

```python
class Book:
    def __init__(self, title, author):
        self.title=title
        self.author=author
    
    def __repr__(self):
        return f"Book('{self.title}', '{self.title}')"

b1=Book("Deep Work", "Cal Newport")
b2=Book("Atomic Habits", "James Clear")

print(b1)
print(b2)

books=[b1, b2]
print(books)
```

4. Create a class `ShoppingCart`:
- attribute: `items` (list)

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Methods:

- `add_item(item_name)`
- `remove_item(item_name) (only if present)`
- `__str__()` → display all items neatly
- `__len__()` → return number of items

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Tasks:

- Add multiple items
- Remove one
- Print cart
- Use `len(cart)`

```python
class ShoppingCart:
    def __init__(self):
        self.items=[]
    
    def add_item(self, item_name):
        self.items.append(item_name)
        return f"Item added: {item_name}"
    
    def remove_item(self, item_name):
        if item_name in self.items:
            self.items.remove(item_name)
            return f"Item removed: {item_name}"
        else:
            return f"Item {item_name} not found"
    
    def __str__(self):
        return f"Cart items: "+",".join(self.items)
    
    def __len__(self):
        return len(self.items)

cart1=ShoppingCart()
cart1.add_item("banana")
cart1.add_item("Apple")
cart1.add_item("Orange")

print(cart1)

cart1.remove_item("banana")

print(cart1)
print(f"Number of items: {len(cart1)}")
```
