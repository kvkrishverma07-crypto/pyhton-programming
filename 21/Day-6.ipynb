2. Modify `Account`:

- Keep `__balance` private
- Create:
    - `get_balance()`
    - `deposit(amount)`
    - `withdraw(amount)` with validation

👉 Ensure balance cannot be accessed or modified directly

```python
class Account:
    def __init__(self, name, pin):
        self.name=name
        self.__pin=pin
        self.__balance=0
    
    def get_balance(self, pin):
        if pin==self.__pin:
            return self.__balance
    
    def deposit(self, amount, pin):
        if pin==self.__pin:
            self.__balance+=amount
            return "Amount deposited successfully!!"
        else:
            return "Incorrect PIN!!"
    
    def withdraw(self, amount, pin):
        if pin!=self.__pin:
            return "Incorrect PIN!!"
        if amount>self.__balance:
            return "Insufficient balance"
        self.__balance-=amount
        return "Amount withdrawed successfully"

obj=Account("Ashish", 2704)
print(obj.deposit(92000, 2704))
print(obj.withdraw(84000, 2704))
print(obj.get_balance(2704))
```

3. Create a class `Shape`:
- method `area()` → just pass (or raise error)

Create child classes:

- `Rectangle`
- `Circle`

👉 Each should implement its own `area()`

```python
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length=length
        self.width=width
    
    def area(self):
        return self.length*self.width

class Circle(Shape):
    def __init__(self, radius):
        self.radius=radius
    
    def area(self):
        return 3.14*self.radius*self.radius

obj1=Rectangle(4, 5)
obj2=Circle(23)

print(obj1.area())
print(obj2.area())
```

4. Create a class `ATM`:

- private variable: `__balance`

👉 Methods:

- `check_balance()`
- `deposit(amount)`
- `withdraw(amount)` (with proper validation)

👉 Constraints:

- No direct balance access
- No negative transactions

```python
class ATM:
    def __init__(self, balance):
        self.__balance=balance
    
    def check_balance(self):
        return self.__balance
    
    def deposit(self, amount):
        if amount<=0:
            return "Not a valid amount"
        else:
            self.__balance+=amount
            return "Deposited successfully"
    
    def withdraw(self, amount):
        if amount<=0 or amount>self.__balance:
            return "Invalid amount"
        else:
            self.__balance-=amount
            return "Withdrawed successfully"

obj=ATM(200000)
print(obj.deposit(84572))
print(obj.withdraw(96542))
print(obj.check_balance())
```
