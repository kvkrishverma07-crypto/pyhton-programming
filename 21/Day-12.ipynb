1. Create a class `Book`:

- attributes: `title`, `days_borrowed`

👉 Method:

- `calculate_late_fee()`:
    - ₹5 per day if `days_borrowed > 7`
    - otherwise 0

👉 Task:

- Create 2 books (one late, one not)
- Print late fees

```python
class Book:
    def __init__(self, title, days_borrowed):
        self.title=title
        self.days_borrowed=days_borrowed
    
    def calculate_late_fee(self):
        if self.days_borrowed>7:
            print("Late Fees: ", 5*(self.days_borrowed-7))
        else:
            print("Late Fees: 0")

b1=Book("ABC", 5)
b2=Book("XYZ", 10)
b1.calculate_late_fee()
b2.calculate_late_fee()
```

2. Create class `User`:

- private `__password`

👉 Methods:

- `set_password(pwd)`
- `check_password(pwd)` → returns True/False

👉 Task:

- Set password
- Check with correct and wrong input

```python
class User:
    def __init__(self):
        self.__password=None
    
    def set_password(self, pwd):
        self.__password=pwd
    
    def check_password(self, pwd):
        print("Password: ", self.__password==pwd)

u1=User()
u2=User()
u1.set_password(123)
u2.set_password(789)
u1.check_password(256)
u2.check_password(789)
```

3. Create:

- class `Item` → `name`, `price`
- class `Order`:
    - attribute: `items` (list)

👉 Methods:

- `add_item(item)`
- `calculate_total()` → sum of prices

👉 Task:

- Add 3 items
- Print total

```python
class Item:
    def __init__(self, name, price):
        self.name=name
        self.price=price

class Order:
    def __init__(self):
        self.items=[]
    
    def add_item(self, item):
        self.items.append(item)
    
    def calculate_total(self):
        sum=0
        for item in self.items:
            sum+=item.price
        
        return sum
    
i1=Item("Laptop", 55000)
i2=Item("Mobile", 22000)
i3=Item("Mouse", 2000)

o1=Order()
o1.add_item(i1)
o1.add_item(i2)
o1.add_item(i3)

o1.calculate_total()
```

4. Create:

- `Shape` → method `area()` (default 0)
- `Rectangle` → length, width
- `Circle` → radius

👉 Override `area()` properly

👉 Task:

- Store both in list
- Print their areas

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length=length
        self.width=width
    
    def area(self):
        return self.length*self.width
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius=radius
    
    def area(self):
        return 3.14*self.radius*self.radius

shapes=[Rectangle(25, 12), Circle(5)]
for shape in shapes:
    print("Area: ", shape.area())
```
