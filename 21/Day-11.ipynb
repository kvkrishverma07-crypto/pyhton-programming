2. Create class `Product`:

- attributes: `name`, `price`

ЁЯСЙ Add method:

- `is_expensive(other_product)`
- returns True if current product price is higher

ЁЯСЙ Task:

- Compare 2 products

```python
class Product:
    def __init__(self, name, price):
        self.name=name
        self.price=price
    
    def is_expensive(self, other):
        return self.price>other.price

obj1=Product("ABC", 20000)
obj2=Product("XYZ", 10000)
obj1.is_expensive(obj2)
```

3. Create class `Cart`:

- attribute: `items` (list)

ЁЯСЙ Methods:

- `add_item(item)`
- `remove_item(item)` (only if exists)
- `show_items()`

ЁЯСЙ Task:

- Add 3 items
- Remove 1
- Print remaining

```python
class Cart:
    def __init__(self):
        self.items=[]
    
    def add_item(self, item):
        self.items.append(item)
        return f"Added item {item}"
    
    def remove_item(self, item):
        self.items.remove(item)
        return f"Removed item {item}"
    
    def show_items(self):
        print("Items in the cart are: ")
        for item in self.items:
            print(item)
        print("\n")

c1=Cart()
c1.add_item("apple")
c1.add_item("banana")
c1.add_item("Orange")
c1.show_items()
c1.remove_item("apple")
c1.show_items()
```

4. Create:

- class `Employee` тЖТ method `get_bonus()` (default = 0)
- class `Manager` тЖТ bonus = 20% salary
- class `Intern` тЖТ bonus = 5% salary

ЁЯСЙ Task:

- Store different employee types in a list
- Loop and print bonuses

```python
class Employee:
    def __init__(self, name, salary):
        self.name=name
        self.salary=salary
    
    def get_bonus(self):
        return 0

class Manager(Employee):
    def get_bonus(self):
        return 0.2*self.salary

class Intern(Employee):
    def get_bonus(self):
        return 0.05*self.salary

Employees=[Employee("krish", 25000), Manager("harsgh", 29000), Intern("buther", 5000)]
for emp in Employees:
    print(f"{emp.name}, Bonus: {emp.get_bonus()}")
```
