1. Create:

- class `Order`
    - attributes: `items` (list), `total`

👉 Methods:

- `add_item(name, price)`
- `calculate_total()`

Create:

- class `Payment`
    - method `pay(order)`

👉 Task:

- Add items
- Calculate total
- Make payment  

```python
class Order:
    def __init__(self):
        self.items=[]
        self.total=0
    
    def add_item(self, name, price):
        self.items.append((name, price))
    
    def calculate_total(self):
        for item in self.items:
            self.total+=item[1]
        return self.total

class Payment():
    def pay(self, order):
        amount=order.calculate_total()
        print(f"Paid Amount: {amount}")

o1=Order()
o1.add_item("Phone", 22000)
o1.add_item("Laptop", 50000)

o1.calculate_total()

p1=Payment()
p1.pay(o1)
```

2. Create:

- class `User`
    - attribute: `borrowed_books`

👉 Add:

- max 2 books limit

Create:

- class `Library`

👉 Method:

- `borrow_book(user, book_name)`

👉 Task:

- Try borrowing more than limit

```python
class User:
    def __init__(self, name):
        self.name=name
        self.borrowed_books=[]

class Library:
    def borrow_book(self, user, book_name):
        if len(user.borrowed_books)>=2:
            print("Can't borrow more then two books at a time")
        else:
            user.borrowed_books.append(book_name)
            print(f"{user.name} borrowed {book_name}")

u1=User("krish")
l1=Library()
l1.borrow_book(u1, "ABC")
l1.borrow_book(u1, "123")
l1.borrow_book(u1, "XYZ")
```

3. Create class `User`:

- attribute: `name`

👉 Method:

- `send_message(other_user, message)`
    - prints: `"A to B: message"`

👉 Task:

- Simulate conversation between 2 users

```python
class User:
    def __init__(self, name):
        self.name=name
    
    def send_message(self, other_user, message):
        print(f"{self.name} to {other_user.name}: {message}")

u1=User("krih")
u2=User("hARASGH")

u1.send_message(u2, "Hi")
u2.send_message(u1, "Hello")
```

4. Create class `ParkingLot`:

- attributes:
    - `capacity`
    - `occupied`

👉 Methods:

- `park()` → if space available
- `leave()` → decrease occupied
- `status()`

👉 Task:

- Try overfilling parking

```python
class ParkingLot:
    def __init__(self, capacity):
        self.capacity=capacity
        self.occupied=0
    
    def park(self):
        if self.occupied<self.capacity:
            self.occupied+=1
            print("Vehicle Parked!!")
        else:
            print("Space full!!!")

    def leave(self):
        self.occupied-=1
        print("Vehicle left successfully!!")
    
    def status(self):
        print(f"Total Vehicles parked: {self.occupied}")

p1=ParkingLot(2)
p1.park()
p1.park()
p1.park()
p1.leave()
p1.park()
p1.status()
```
