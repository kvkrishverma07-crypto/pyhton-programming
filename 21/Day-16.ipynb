1. Create class `Student`:

- attributes: `name`, `attendance` (initially 0)

👉 Methods:

- `mark_present()` → +1
- `show_attendance()`

👉 Task:

- Mark attendance multiple times
- Print total

```python
class Student:
    def __init__(self, name):
        self.name=name
        self.attendance=0
        
    def mark_present(self):
        self.attendance+=1
        print("Present!!")
    
    def show_attendance(self):
        print(f"Current Attendance: {self.attendance}")

s1=Student("Ashish")
s1.mark_present()
s1.mark_present()
s1.show_attendance()
```

2. Create:

- class `Teacher`
- class `Student`

👉 Teacher method:

- `assign_homework(student, task)`

👉 Student:

- stores tasks in a list

👉 Task:

- Assign 2 tasks to a student
- Print tasks

```python
class Student:
    def __init__(self, name):
        self.name=name
        self.tasks=[]
    
    def add_task(self, task):
        self.tasks.append(task)
    
    def show_tasks(self):
        print("Tasks: ")
        for task in self.tasks:
            print(task, end=" ")

class Teacher:
    def assign_homework(self, student, task):
        student.add_task(task)

t1=Teacher()
s1=Student("Ashish")
t1.assign_homework(s1, "Maths")
t1.assign_homework(s1, "Python")
s1.show_tasks()
```

3. Create class `Account`:

- attributes: `name`, `balance`

👉 Methods:

- `deposit(amount)`
- `withdraw(amount)`
- `transfer(other_account, amount)`

👉 Task:

- Transfer money from one account to another

```python
class Account:
    def __init__(self, name, balance):
        self.name=name
        self.balance=balance
    
    def deposit(self, amount):
        self.balance+=amount
    
    def withdraw(self, amount):
        self.balance-=amount
    
    def transfer(self, other, amount):
        self.balance-=amount
        other.balance+=amount
    
    def check_balance(self):
        print("Current Balance:", self.balance)

a1=Account("Ashish", 22000)
a2=Account("Tanveer", 11000)
a1.deposit(8000)
a2.withdraw(1000)
a1.transfer(a2, 10000)
a1.check_balance()
a2.check_balance()
```

4. Create class `Door`:

- attribute: `is_locked` (True initially)

👉 Methods:

- `lock()`
- `unlock(password)`
    - correct password = `"1234"`
- `status()`

👉 Task:

- Try wrong password
- Then correct password

```python
class Door:
    def __init__(self):
        self.is_locked=True
    
    def lock(self):
        if self.is_locked!=True:
            self.is_locked=True
            print("Locked!!")
        else:
            print("Already Locked!!")
    
    def unlock(self, password):
        if password==1234:
            print("Door Unlocked!!")
            self.is_locked=False
        else:
            print("Wrong Password")
    
    def status(self):
        print("Door Locked: ", self.is_locked)

d1=Door()
d1.lock()
d1.unlock(4567)
d1.unlock(1234)
d1.status()
```
