1. Create a class `Student` with:

- attributes: `name`, `age`
- a method `display()` that prints both values

```python
class Student:
    def set_details(self, name, age):
        self.name=name
        self.age=age
        
    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

details=Student()
details.set_details("krish", 18)
details.display()
```

3. Create a class `Rectangle`:  
- attributes: `length`, `width`
- methods:  
    - `area()` → returns area
    - `perimeter()` → returns perimeter

```python
class Rectangle:
    def __init__(self, length, width):
        self.length=length
        self.width=width
    
    def area(self):
        return self.length*self.width
    
    def perimeter(self):
        return 2*(self.length+self.width)

figure=Rectangle(20, 15)
print(f"Area: {figure.area()}, Perimeter: {figure.perimeter()}")
```

4. Create a class `BankAccount`:
- attributes: `account_holder`, `balance`
- methods:
    - `deposit(amount)`
    - `withdraw(amount)` (only if balance is sufficient)
    - `display_balance()`

```python
class BankAccount:
    def __init__ (self, account_holder):
        self.account_holder=account_holder
        self.balance=0
    
    def deposit(self, amount):
        if amount<=0:
            return "Invalid Amount"
        else:
            self.balance+=amount
    
    def withdraw(self, amount):
        if amount>self.balance:
            return "Insufficient Balance"
        else:
            self.balance-=amount
    
    def display_balance(self):
        return self.balance

Acc1=BankAccount("Ashish")
Acc1.deposit(1000)
Acc1.withdraw(500)
Acc1.display_balance()
```
