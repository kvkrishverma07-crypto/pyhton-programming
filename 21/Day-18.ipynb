1. Create class `CartItem`:

- attributes: `name`, `price`, `quantity`

Create class `Cart`:

- attribute: `items` (list)

👉 Methods:

- `add_item(name, price, quantity)`
- `calculate_total()` → price × quantity

👉 Task:

- Add multiple items
- Print total

```python
class CartItem:
    def __init__(self, name, price, quantity):
        self.name=name
        self.price=price
        self.quantity=quantity

class Cart:
    def __init__(self):
        self.items=[]
    
    def add_item(self, name, price, quantity):
        item=CartItem(name, price, quantity)
        self.items.append(item)
    
    def calculate_total(self):
        total=0
        for item in self.items:
            total+=item.price*item.quantity
        print("Total: ", total)

cart=Cart()
cart.add_item("Laptop", 22000, 2)
cart.add_item("Phone", 5000, 4)
cart.calculate_total()
```

2. Create:

- class `User`
    - attribute: `name`
- class `Notification`

👉 Method:

- `send(user, message)`
    - prints: `"Notification to Ashish: message"`

👉 Task:

- Send multiple notifications

```python
class User:
    def __init__(self, name):
        self.name=name

class Notification:
    def send(self, user, message):
        print(f"Notification to {user.name}: {message}")

u1=User("Ashish")
n1=Notification()
n1.send(u1, "How are you?")
```

3. Create class `Student`:

- attributes:
    - `name`
    - `marks` (list)

👉 Methods:

- `add_mark(mark)`
- `is_passed()` → pass if average ≥ 40

👉 Task:

- Add marks
- Check pass/fail

```python
class Student:
    def __init__(self, name):
        self.name=name
        self.marks=[]
    
    def add_mark(self, mark):
        self.marks.append(mark)
    
    def is_passed(self):
        if len(self.marks)==0:
            return False
        avg=sum(self.marks)/len(self.marks)
        if avg>=40:
            print("PASSED!!")
        else:
            print("FAILED!!!")

s1=Student("Ashish")
s2=Student("Tanveer")

s1.add_mark(40)
s1.add_mark(70)
s1.is_passed()

s2.add_mark(50)
s2.add_mark(40)
s2.add_mark(5)
s2.is_passed()
```

4. Create class `Elevator`:

- attributes:
    - `current_floor`
    - `max_floor`

👉 Methods:

- `go_up()` → increase floor (within limit)
- `go_down()` → decrease floor (not below 0)
- `status()`

👉 Task:

- Try going beyond limits

```python
class Elevator:
    def __init__(self, max_floor):
        self.current_floor=0
        self.max_floor=max_floor
    
    def go_up(self):
        if self.current_floor>=self.max_floor:
            print("Already at top floor!!")
        else:
            self.current_floor+=1
            print("Going Up")
    
    def go_down(self):
        if self.current_floor<=0:
            print("Already at bottom floor!!")
        else:
            self.current_floor-=1
            print("Going Down!!")
    
    def status(self):
        print(f"Current Floor: {self.current_floor}")

e1=Elevator(3)
e1.go_down()
e1.go_up()
e1.go_up()
e1.go_up()
e1.go_up()
e1.go_down()
e1.status()
```
