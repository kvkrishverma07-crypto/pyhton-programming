1. Create class `Timer`:

- attribute: `seconds` (initially 0)

👉 Methods:

- `tick()` → increases time by 1 second
- `reset()` → sets time to 0
- `show_time()` → prints current seconds

👉 Task:

- Call `tick()` multiple times
- Reset and check

```python
class Timer:
    def __init__(self):
        self.seconds=0
    
    def tick(self):
        self.seconds+=1
        print("Tick!!")
    
    def reset(self):
        self.seconds=0
        print("BOOM!!!")
    
    def show_time(self):
        return f"Current Timer: {self.seconds}"

t1=Timer()
t1.tick()
t1.tick()
t1.reset()
t1.tick()
t1.tick()
t1.show_time()
```

2. Create class `LoginSystem`:

- attributes:
    - `username`
    - `attempts` (initially 0)

👉 Methods:

- `login(password)`
    - assume correct password = `"admin123"`
    - if wrong → increase attempts
    - if attempts ≥ 3 → block login

👉 Task:

- Try wrong password multiple times

```python
class LoginSystem:
    def __init__(self, username):
        self.username=username
        self.attempts=0
    
    def login(self, password):
        if self.attempts>=3:
            print("Account Blocked!!")
            return
        else:
            if password=="admin123":
                print("Login Successful")
                self.attempts=0
            else:
                print("Wrong Password!!")
                self.attempts+=1
    
u1=LoginSystem("Ashish")
u1.login("ABC")
u1.login("123")
u1.login("246")
u1.login(546)
```

3. Create class `Playlist`:

- attribute: `songs` (list)

👉 Methods:

- `add_song(song)`
- `remove_song(song)`
- `play_all()` → prints all songs

👉 Task:

- Add 3 songs
- Remove 1
- Play remaining

```python
class Playlist:
    def __init__(self):
        self.songs=[]
    
    def add_song(self, song):
        self.songs.append(song)
        print(f"Song {song} added successfully")
    
    def remove_song(self, song):
        self.songs.remove(song)
        print(f"Song {song} removed successfully")
    
    def play_all(self):
        for song in self.songs:
            print(f"Currently playing: {song}")

p1=Playlist()
p1.add_song("song1")
p1.add_song("song2")
p1.add_song("song3")
p1.remove_song("song2")
p1.play_all()
```

4. Create class `Wallet`:

- private `__balance`

👉 Methods:

- `add_money(amount)`
- `spend_money(amount)` (only if enough balance)
- `check_balance()`

👉 Task:

- Add money
- Spend valid + invalid amounts

```python
class Wallet:
    def __init__(self):
        self.__balance=0
    
    def add_money(self, amount):
        self.__balance+=amount
        print(f"Money added: {amount} inr")
    
    def spend_money(self, amount):
        if amount>self.__balance:
            print("Insufficient Balance!!!")
            return
        else:
            self.__balance-=amount
            print(f"Money spent: {amount} inr")
    
    def check_balance(self):
        print(f"Current Balance: {self.__balance} inr")

w1=Wallet()
w1.add_money(25000)
w1.spend_money(13000)
w1.spend_money(13000)
w1.check_balance()
```
