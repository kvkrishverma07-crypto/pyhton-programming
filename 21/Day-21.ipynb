`System` (Main controller)  

* attribute:

  * list of all courses

👉 methods:

* `add_course(course)`
* `enroll_student(student, course_name)`
* `show_course_details(course_name)`

---

### ⚙️ **Rules (Important)**

* A student should NOT be enrolled twice in same course
* Course must exist before enrolling
* Use object interaction (no direct data manipulation)

---

### 🧪 **What You Should Demonstrate**

Your code should:

1. Create students
2. Create courses
3. Enroll students
4. Prevent duplicate enrollment
5. Show course details

---

```python
class Student:
    def __init__(self, name):
        self.name=name
        self.courses=[]
    
    def enroll(self, course):
        if course not in self.courses:
            self.courses.append(course)
        else:
            print(f"{self.name} already enrolled in {course.name}")
    
    def show_course(self):
        print("Courses of", self.name)
        for course in self.courses:
            print(course.name, end=" ")

class Course:
    def __init__(self, name):
        self.name=name
        self.students=[]
    
    def add_student(self, student):
        if student not in self.students:
            self.students.append(student)
        else:
            print(f"{student.name} already in {self.name}")
    
    def show_students(self):
        print("Students enrolled in", self.name, ":")
        for student in self.students:
            print(student.name, end=" ")

class System:
    def __init__(self):
        self.courses=[]
    
    def add_course(self, course):
        if course not in self.courses:
            self.courses.append(course)
        else:
            print(f"{course.name} already added!!")
    
    def enroll_student(self, student, course_name):
        for course in self.courses:
            if course.name==course_name:
                if student in course.students:
                    print(f"{student.name} already enrolled in {course.name}")
                    return
                
                course.add_student(student)
                student.enroll(course)
                
                print(f"{student.name} enrolled in {course.name}")
                return
            
        print("Course not found")
    
    def show_course_details(self, course_name):
        for course in self.courses:
            if course.name==course_name:
                course.show_students()
                return
        print("Course not found")

s1=Student("KRISH")
s2=Student("HARSGH")

c1=Course("Python")
c2=Course("DSA")

system=System()

system.add_course(c1)
system.add_course(c2)

system.enroll_student(s1, "Python")
system.enroll_student(s1, "Python")
system.enroll_student(s2, "Python")
system.enroll_student(s2, "DSA")

system.show_course_details("Python")
```
