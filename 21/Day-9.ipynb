1. Create a class `Engine`:

- method: `start()`

Create a class `Car`:

- instead of inheriting, use **Engine** as an attribute

👉 Task:

- Call engine’s method using car object

````python
class Engine:
    def start(self):
        print("Engine Started!!")

class Car:
    def __init__(self):
        self.engine=Engine()
    
    def drive(self):
        self.engine.start()
        return "Car is moving"

obj1=Car()
obj1.drive()
````

2. Create:

👉 Now ask yourself:

- Is inheritance correct here?

Then create:

- `Heart` class → method `pump()`
- `Human` class → uses `Heart` via composition

👉 Compare both designs     

````python
class Animal:
    def eat(self):
        print("Animal is eating!!")

class Dog(Animal):
    pass

class Heart:
    def pump(self):
        print("Heart is pumping!!")

class Human:
    def __init__(self):
        self.heart=Heart()
    
    def live(self):
        self.heart.pump()
        print("Human is living")

obj1=Dog()
obj1.eat()

obj2=Human()
obj2.live()
````

4. Create:

- class `Battery` → method `charge()`
- class `Phone`

👉 Phone should:

- use Battery (composition)
- method `power_on()` should depend on battery

````python
class Battery:
    def charge(self):
        print("Battery fully charged!!")

class Phone:
    def __init__(self):
        self.battery=Battery()
    
    def power_on(self):
        self.battery.charge()
        print("Phone powered on!!")

obj1=Phone()
obj1.power_on()
````

5. Create:

- class `Employee` (base class)
- class `Developer` (inherits Employee)

👉 Also create:

- class `Laptop`

👉 Developer should:

- inherit from Employee
- also “have a” Laptop

````python
class Laptop:
    def code(self):
        print("Coding...")

class Employee:
    def __init__(self, name):
        self.name=name

class Developer(Employee):
    def __init__(self, name):
        super().__init__(name)
        self.laptop=Laptop()
    
    def work(self):
        print(f"{self.name} is working!!")
        self.laptop.code()

obj1=Developer("krish")
obj1.work()
````
