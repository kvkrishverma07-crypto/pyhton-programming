1. Create class `User`:

- attribute: `username`

👉 Method:

- `is_valid()`:
    - length ≥ 5
    - no spaces

👉 Task:

- Test valid and invalid usernames

```python
class User:
    def __init__(self, username):
        self.username=username
    
    def is_valid(self):
        if len(self.username)>=5 and " " not in self.username:
            print("VALID!!")
        else:
            print("INVALID!!")

u1=User("Ashish")
u2=User("Tan veer")

u1.is_valid()
u2.is_valid()
```

2. Create class `Timer`:

- attribute: `seconds`

👉 Methods:

- `add_time(sec)`
- `show_time()` → print in format: `mm:ss`

👉 Example:

125 seconds → `2:05`

```python
class Timer:
    def __init__(self):
        self.seconds=0
    
    def add_time(self, sec):
        if sec>0:
            self.seconds+=sec
        else:
            print("Invalid seconds")
    
    def show_time(self):
        minutes=self.seconds//60
        seconds=self.seconds%60
        print(f"{minutes}:{seconds:02d}")

t1=Timer()

t1.add_time(125)
t1.show_time()
```

3. Create:

- class `Candidate` → `name`, `votes`
- class `Election` → list of candidates

👉 Methods:

- `add_candidate(name)`
- `vote(name)`
- `show_results()`

👉 Task:

- Vote multiple times
- Print results

```python
class Candidate:
    def __init__(self, name):
        self.name=name
        self.votes=0
    
class Election:
    def __init__(self):
        self.candidates=[]
    
    def add_candidate(self, name):
        self.candidates.append(name)
    
    def vote(self, name):
        for candidate in self.candidates:
            if candidate.name==name:
                candidate.votes+=1
                return
        print("Candidate not found!!")
    
    def show_results(self):
        for candidate in self.candidates:
            print(f"{candidate.name}: Votes={candidate.votes}")

c1=Candidate("KRISH")
c2=Candidate("HASRG")

e1=Election()
e1.add_candidate(c1)
e1.add_candidate(c2)

e1.vote("KRISH")
e1.vote("KRISH")
e1.vote("KRISH")
e1.vote("HARSG")

e1.show_results()    
```

4. Create class `Light`:

- attribute: `is_on` (False)

👉 Methods:

- `toggle()` → switch state
- `status()`

👉 Task:

- Toggle multiple times
- Print status

```python
class Light:
    def __init__(self):
        self.is_on=False
    
    def toggle(self):
        self.is_on=not self.is_on
    
    def status(self):
        print(f"Torch on: {self.is_on}")

l1=Light()

l1.toggle()
l1.status()
l1.toggle()
l1.status()
l1.toggle()
l1.status()
```
