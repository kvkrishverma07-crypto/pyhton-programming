1. Create a class `Vector`:

- attributes: `x`, `y`

đź‘‰ Overload `+` operator:

- Adding two vectors should return a new vector
Example:
```python
v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2   # (6, 8)
```

````python
class Vector:
    def __init__(self, x, y):
        self.x=x
        self.y=y
    
    def __add__(self, other):
        return Vector(self.x+other.x, self.y+other.y)

v1=Vector(2, 3)
v2=Vector(5, 6)
v3=v1+v2
print(v3.x, v3.y)
````

2. Create a class `Product`:

- attributes: `name`, `price`

đź‘‰ Overload `<` operator:

- Compare based on price

Example:
```python
p1 < p2   # True if p1.price < p2.price
```

````python
class Product:
    def __init__(self, name, price):
        self.name=name
        self.price=price
    
    def __lt__(self, other):
        if self.price<other.price:
            return True
        else:
            return False

obj1=Product("Phone", 25000)
obj2=Product("AC", 35000)
print(obj1<obj2)
````

3. Enhance `Vector` class:

đź‘‰ Add:

- `__str__()` â†’ display as `(x, y)`

đź‘‰ Then:

- Print result of addition nicely

````python
class Vector:
    def __init__(self, x, y):
        self.x=x
        self.y=y
    
    def __add__(self, other):
        return Vector(self.x+other.x, self.y+other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"

v1=Vector(10, 5)
v2=Vector(30, 90)
print(v1+v2)
````

4. Python doesnâ€™t support traditional overloading.  

Create class `Calculator`:

đź‘‰ Method `add()` should:

- work with 2 numbers
- work with 3 numbers
- handle dynamically using default arguments or `*args`

Example:
```python
add(2, 3)
add(2, 3, 4)
```

````python
class Calculator:
    def add(Self, *args):
        return sum(args)

c1=Calculator()
print(c1.add(2, 3))
print(c1.add(2, 3, 4))
print(c1.add(2, 3, 4, 5))
````

5. Create class `CartItem`:

- attributes: `name`, `price`, `quantity`

đź‘‰ Implement:

- `__add__()` â†’ combine same items (increase quantity)
- `__str__()` â†’ display nicely

Example:
```python
item1 + item2
```

````python
class CartItem:
    def __init__(self, name, price, quantity):
        self.name=name
        self.price=price
        self.quantity=quantity
    
    def __add__(self, other):
        if self.name==other.name:
            return CartItem(self.name, self.price, self.quantity+other.quantity)
        else:
            return "Can't combine diff items!!"
    
    def __str__(self):
        return f"{self.name}- Price: {self.price}, Quantity: {self.quantity}"

item1=CartItem("Phone", 25000, 5)
item2=CartItem("Phone", 25000, 7)
item3=CartItem("Laptop", 50000, 2)

print(item1+item2)
print(item1+item3)
````
