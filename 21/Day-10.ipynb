1. Design a simple system:  
**Classes:**
- `Book`
    - attributes: `title`, `author`, `is_available`
- `Library`
    - stores list of books  
    
**Methods:**
- add_book(book)
- borrow_book(title)
- return_book(title)

👉 Task:

- Add multiple books
- Borrow one
- Try borrowing again (handle properly)

```python
class Book:
    def __init__(self, title, author):
        self.title=title
        self.author=author
        self.is_available=True

class Library:
    def __init__(self):
        self.books=[]
        
    def add_book(self, book):
        self.books.append(book)
        print("Book added successfully!!")
    
    def borrow_book(self, title):
        for book in self.books:
            if title==book.title:
                if book.is_available:
                    book.is_available=False
                    print(title, "Borrowed!!")
                    return
                else:
                    print("Book not Available!!")
                    return
        else:
            print("Book not Found!!")
    
    def return_book(self, title):
        for book in self.books:
            if title==book.title:
                book.is_available=True
                print(title+" returned successfully!!")
                return
        else:
            print("Book Not Found!!")

b1=Book("book1", "Author1")
b2=Book("book2", "author2")
l1=Library()
l1.add_book(b1)
l1.add_book(b2)
l1.borrow_book("book1")
l1.borrow_book("book2")
l1.return_book("book1")
l1.borrow_book("book2")
    
```

2. Extend previous system:  
Add class `User`:  
- attribute: `name`  
  
👉 Modify:
  
- When a book is borrowed → track **which user borrowed it**

```python
class User:
    def __init__(self, name):
        self.name=name
        self.borrowed_books=[]

class Book:
    def __init__(self, title, author):
        self.title=title
        self.author=author
        self.is_available=True

class Library:
    def __init__(self):
        self.books=[]
        
    def add_book(self, book):
        self.books.append(book)
        print("Book added successfully!!")
    
    def borrow_book(self, title, user):
        for book in self.books:
            if title==book.title:
                if book.is_available:
                    book.is_available=False
                    user.borrowed_books.append(title)
                    print(f"{title} borrowed by {user.name}!!")
                    return
                else:
                    print("Book not Available!!")
                    return
        else:
            print("Book not Found!!")
    
    def return_book(self, title, user):
        for book in self.books:
            if title==book.title:
                book.is_available=True
                user.borrowed_books.remove(title)
                print(title+" returned successfully by "+user.name)
                return
        else:
            print("Book Not Found!!")

b1=Book("book1", "Author1")
b2=Book("book2", "author2")

u1=User("Ashish")
l1=Library()
l1.add_book(b1)
l1.add_book(b2)
l1.borrow_book("book1", u1)
l1.borrow_book("book2", u1)
l1.return_book("book1", u1)
l1.borrow_book("book2", u1)
    
```
