1. Create class `Temperature`:

- attribute: `celsius`

đź‘‰ Methods:

- `to_fahrenheit()`
- `to_kelvin()`

đź‘‰ Task:

- Convert and print both values

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius=celsius
    
    def to_fahrenheit(self):
        return (self.celsius*9/5)+32
    
    def to_kelvin(self):
        return self.celsius+273.15

t1=Temperature(32)
print(t1.to_fahrenheit())
print(t1.to_kelvin())
```

3. Create class `Product`:

- attributes: `name`, `stock`

đź‘‰ Methods:

- `sell(quantity)`
    - only if enough stock
- `restock(quantity)`

đź‘‰ Task:

- Try:
    - valid sell
    - invalid sell (more than stock)

```python
class Product:
    def __init__(self, name, stock):
        self.name=name
        self.stock=stock
    
    def sell(self, quantity):
        if quantity<=self.stock:
            self.stock-=quantity
            print("Sold Stocks:", quantity)
        else:
            print("Not enough stocks!!")
    
    def restock(self, quantity):
        self.stock+=quantity
        print("Stock added:", quantity)

p1=Product("Phone", 25)
p1.sell(30)
p1.restock(15)
p1.sell(30)
```

4. Create class `Student`:

- attributes: `name`, `marks` (list)

đź‘‰ Methods:

- `add_mark(mark)`
- `average_marks()`
- `get_grade()`:
    - â‰Ą90 â†’ A
    - â‰Ą75 â†’ B
    - â‰Ą50 â†’ C
    - else â†’ F

đź‘‰ Task:

- Add marks
- Print average + grade

```python
class Student:
    def __init__(self, name):
        self.name=name
        self.marks=[]
    
    def add_mark(self, mark):
        self.marks.append(mark)
        print("Marks Added:", mark)
    
    def average_marks(self):
        self.avg=sum(self.marks)/len(self.marks)
        print("Average:", self.avg)
    
    def get_grade(self):
        if self.avg>=90:
            print("Grade: A")
        elif self.avg>=75:
            print("Grade: B")
        elif self.avg>=50:
            print("Grade: C")
        else:
            print("Grade: F")

s1=Student("Ashish")
s1.add_mark(80)
s1.add_mark(95)
s1.add_mark(80)
s1.add_mark(100)
s1.add_mark(76)
s1.add_mark(88)
s1.average_marks()
s1.get_grade()
```
