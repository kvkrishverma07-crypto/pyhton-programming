```python

name = "Krish"
sap_id = 590021464
dob = "01 Oct 2007"
programme = "AI & ML"
semester = 2
print(f"NAME: {name}")
print(f"SAP ID: {sap_id}")
print(f"DATE OF BIRTH: {dob}")
print(f"""ADDRESS: UPES 
                 Bidholi Campus 
                 Pincode: 248007""")

print(f"Programme: {programme}")
print(f"Semester: {semester}")
```

## Why we used these specific types:                                    
name & dob (Strings): We use quotes because these contain letters and spaces.                                        

sap_id (Integer): Since a SAP ID is a whole number, we don't strictly need quotes (though you could use them if you don't plan to do math with the ID).                        

semester (Integer): A simple number.                                                   

f"..." (f-strings): This is the cleanest way to insert your variables directly into your labels (like NAME: or SAP ID:).                                

Triple Quotes ("""): This allowed us to indent "Bidholi Campus" and "Pincode" exactly where you wanted them, just by pressing Enter and Space inside the code.
