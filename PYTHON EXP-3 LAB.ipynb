## 6.	Find whether a given year is a leap year or not. 

```python
year = int(input("Enter any year"))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap year")
else:
    print("Not leap year" )
```

Why start with 400?           
Because if a year like 2000 comes in, it is divisible by 4, 100, and 400. If we checked year % 4 first, the code would stop there and say "Leap Year," which is correct, but for a year like 1900, it would also stop at year % 4 even though 1900 is not a leap year (because it's divisible by 100 but not 400).

By checking the most specific rule (400) first, we ensure the logic is foolproof.

##  7. Write a program which takes any date as input and display next date of the calendar

```python
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    days_in_month[1] = 29
day += 1
if day > days_in_month[month - 1]:
    day = 1
    month += 1
    if month > 12:
        month = 1
        year += 1
print("Next date is:", day, "/", month, "/", year)
```
